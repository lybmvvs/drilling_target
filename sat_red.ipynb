```python
import pandas as pdd
import statistics
por_1 = pdd.read_excel('2_task_sat_BP_12_1.xlsx')
por_2 = pdd.read_excel('2_task_sat_BP_12_2.xlsx')
por_3 = pdd.read_excel('2_task_sat_BP_12_3.xlsx')
summ=pdd.concat([por_1,por_2,por_3])
por_4 = pdd.read_excel('2_task_sat_U_1_1.xlsx')
por_5 = pdd.read_excel('2_task_sat_U_1_2.xlsx')
summ1=pdd.concat([por_4,por_5])
summ1
```

```python
summ = summ.groupby('Well').agg(
    {'Oil saturation, %': lambda x: 
     statistics.mean(x)}
)
summ
```

```python
summ.reset_index(inplace=True)

```

```python
cordinates = pdd.read_excel('new_well_info.xlsx',header=[0],index_col=[0])
cordinates=cordinates.drop(['Thickness, meters','Effective thickness, meters','Depth, meters'],axis=1)
cordinates.reset_index(inplace=True)
cordinates
```

```python
final_poro = summ.merge(cordinates, on='Well')
final_poro
```

```python
final_poro.to_excel('2_TASK_sat_BP_12.xlsx')
```
