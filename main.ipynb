```python
import numpy as np
import pandas as pd
import datetime
import random
import missingno as msno

```

```python
production = pd.read_excel('production.xlsx',skiprows=2)

old_heading = list(production.columns)

units = list(production.loc[0])

new_heading = [
    
    str(name).capitalize() if str(unit) == 'nan' else str(name).capitalize() + ', ' +  str(unit)  for name, unit in  dict(zip(old_heading, units)).items()
]

production.columns = new_heading

production = production.drop([0], axis = 0).reset_index(drop=True)

production
```

```python
data1 = production.drop(['Reservoir','Well type','Working time, h','Idle time, h','Oil production rate, t/day','Liquid production rate, t/day','Water production rate, t/day'],axis=1)
data2=data1.drop(['Oil production, tonne','Liqud production (volume), m3','Liqud production , tonne','Water production , tonne','Gas production , m3'],axis=1)
data2 = data2.drop(data2[data2['Oil production (volume), m3'] == 0].index)
data2 = data2.drop(data2[(data2.Date != 1.2015)&(data2.Date != 2.2015)&(data2.Date != 3.2015)&(data2.Date != 4.2015)&(data2.Date != 5.2015)&(data2.Date != 6.2015)&(data2.Date != 7.2015)&(data2.Date != 8.2015)&(data2.Date != 9.2015)&(data2.Date != 10.2015)&(data2.Date != 11.2015)&(data2.Date != 12.2015)&(data2.Date != 1.2016)&(data2.Date != 2.2016)&(data2.Date != 3.2016)&(data2.Date != 4.2016)&(data2.Date != 5.2016)&(data2.Date != 6.2016)&(data2.Date != 7.2016)&(data2.Date != 8.2016)&(data2.Date != 9.2016)&(data2.Date != 10.2016)&(data2.Date != 11.2016)&(data2.Date != 12.2016)&(data2.Date != 1.2017)&(data2.Date != 2.2017)&(data2.Date != 3.2017)&(data2.Date != 4.2017)&(data2.Date != 5.2017)&(data2.Date != 6.2017)&(data2.Date != 7.2017)&(data2.Date != 8.2017)&(data2.Date != 9.2017)&(data2.Date != 10.2017)&(data2.Date != 11.2017)&(data2.Date != 12.2017)].index)
data2 = data2.drop(data2[data2['Watercut, %'] >= 70.0].index)
data2
#удаляю ненужные столбцы, удаляю скважины с 0-й добычей, удаляю все года, кроме 2015,2016,2017
```

```python
production_per_well = data2.groupby('Well').agg(
    {'Oil production (volume), m3': lambda x: 
     sum(x)}
)
production_per_well
#суммарные объемы за 3 года
```

```python

data2['Год'] = data2.apply(
    lambda x: 
    2017 if '2017' in str(x['Date'])
    else 2016 if '2016' in str(x['Date'])
    else 2015, axis=1)
data2
#добавляю столбец с соответствующим годом
```

```python
production_per_well = data2.groupby(by=['Well','Год']).agg(
    {'Oil production (volume), m3': lambda x: 
     sum(x)}
)
production_per_well
#считаю добычу за каждый год для каждой скважины, если она 0, то она не отображается
```

```python
production_per_well.reset_index(inplace=True)#мультииндекс в столбец
production_per_well
```

```python
production_per_well1 = production_per_well.groupby('Well').agg(
    {'Oil production (volume), m3': lambda x: 
     x.tolist()}
)
production_per_well1
#перевожу датафреймы в список для каждой скважины, в нем 1-3 значения добычи за каждый из годов
```

```python

production_per_well1['Количество'] = production_per_well1.apply(
    lambda x: 
    3 if len(x['Oil production (volume), m3'])==3
    else 0, axis=1)

production_per_well1
#вычисляю число лет. за которые есть данные (для простоты: 3 если есть за все года, 0 - если хоть за один из годов нету)
```

```python
production_per_well1 = production_per_well1.drop(production_per_well1[production_per_well1['Количество'] != 3].index)
production_per_well1
#удаляю те, в которых нет добычи за каждый из трех годов
```

```python
production_per_well1 = production_per_well1.drop('Количество',axis=1)
production_per_well1
```

```python
production_per_well1.reset_index(inplace=True)
#мультииндекс в столбец
a=production_per_well1['Well'].tolist()
a
#список с рабочими скважинами
```

```python
production_per_well=production_per_well[production_per_well.Well.isin(a)].reset_index(drop=True)

production_per_well
#удаляем те, которых нет в списке выше(у которых есть нули за какой-нибудь год)
```
